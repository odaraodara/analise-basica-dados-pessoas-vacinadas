### 3. Crie uma nova coluna chamada "tipo_faixa_etaria", onde deverá conter os seguintes valores:

 - Crianças: 3 a 12 anos

 - Adolescentes: 13 a 19 anos

 - Jovens adultos: 20 a 39 anos

 - Adultos de meia-idade: 40 a 59 anos

 - Idosos: 60 anos ou mais

```python
# apagando valores nulos da coluna "idade":
df_copy = df.dropna(subset="idade")
```

```python
# conferindo que não tem valores nulos na coluna:
df_null = df_copy.isnull().sum()
df_null
```

```python
df_copy['idade'].dtype
```

```python
# transformando os valores para valores numéricos
df_copy['idade'] = pd.to_numeric(df_copy['idade'], errors='coerce', downcast='integer')
```

```python
def filtra_tipo_idade(idade):
    if idade > 59:
        return "Idosos"
    elif idade > 39:
        return "Adultos de meia-idade"
    elif idade > 19:
        return "Jovens adultos"
    elif idade > 12:
        return "Adolescentes"
    elif idade > 3:
        return "Crianças"
    else:
        return "Não definido"
```

```python
df_copy['tipo_faixa_etaria'] = df_copy['idade'].apply(lambda x:filtra_tipo_idade(x))

# organizar ordem das colunas
df_copy = df_copy[['faixa_etaria','tipo_faixa_etaria', 'idade', 'sexo', 'raca_cor', 'municipio', 'grupo',
       'categoria', 'lote', 'vacina_fabricante', 'descricao_dose', 'cnes',
       'sistema_origem', 'data_vacinacao']]

df_copy.head(5)
```

### 4. Crie uma nova coluna chamada "regiao", onde deverá conter valores:

 - Metropolitana: Recife, Olinda, Jaboatão dos Guararapes, Paulista, Camaragibe, São Lourenço da Mata, Abreu e Lima, Igarassu, Cabo de Santo Agostinho

 - Interior: Cidades que não estão inclusas em Metropolitana

```python
metropolitana = 'Recife, Olinda, Jaboatão dos Guararapes, Paulista, Camaragibe, São Lourenço da Mata, Abreu e Lima, Igarassu, Cabo de Santo Agostinho'
metropolitana = metropolitana.upper().split(', ')
metropolitana
```

```python
def acharRegiao (municipio):
  if municipio in metropolitana:
    return  'METROPOLITANA'
  else:
    return "INTERIOR"
```

```python
df_copy['regiao'] = df_copy['municipio'].apply(lambda x:acharRegiao(x))
```

```python
df_copy.columns
```

```python
# organizar ordem das colunas
df_copy = df_copy[['faixa_etaria', 'tipo_faixa_etaria', 'idade', 'sexo', 'raca_cor',
       'municipio','regiao', 'grupo', 'categoria', 'lote', 'vacina_fabricante',
       'descricao_dose', 'cnes', 'sistema_origem', 'data_vacinacao']]
```
